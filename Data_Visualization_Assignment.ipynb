##question 4. Using the given x and y data, generate a 3D surface plot to visualize the function

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
fig = go.Figure()
fig.add_trace(go.Surface(x = df['X'], y = df['Y'], z = df['Z'], colorscale = 'Viridis'))
```

# **BOKEH ASSIGNMENT:**

##question 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title = 'sine wave', x_axis_label = 'x', y_axis_label = 'y')
p.line(x, y, line_width = 2)
show(p)
```

##question 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
x =np.random.randn(100)
y =np.random.randn(100)
sizes = np.random.randint(10, 50, 100)
p = figure(title = 'scatter plot', x_axis_label = 'x', y_axis_label = 'y')
p.scatter(x, y, size = sizes, color = 'blue')
show(p)
```

##question 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
p = figure(x_range = fruits, title = 'fruit counts', x_axis_label = 'fruits', y_axis_label = 'counts')
p.vbar(x = fruits, top = counts, width = 0.5)
show(p)
```

##question 4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
p = figure(title = 'histogram', x_axis_label = 'value', y_axis_label = 'frequency')
p.quad(top = hist, bottom = 0, left = edges[:-1], right = edges[1:], fill_color = 'skyblue', line_color = 'black')
show(p)
```
